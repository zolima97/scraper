```python
start = time.time()
lis = []

for i in links_ready[100:]:
    email_scraped = driver.get(str(i))
    time.sleep(5)
    tag_name = driver.find_elements_by_tag_name('a') 
    info_true = [i.get_attribute('href') for i in tag_name]
    
    for elem in info_true:
        if elem != None:
            if "mailto" in elem:
                lis.append(elem.format(i))
            else:
                continue
        else:
            lis.append("Manca:{}".format(i))

    
print(f"Time: {time.time() - start}")
```

```python
def_lis = []
for elem in lis:
    element = elem.replace("mailto:", "")
    def_lis.append(element)
```

```python
def_lis = list( dict.fromkeys(def_lis) )
print(def_lis)
```

```python
#open the email file
textfile = open("e-mail.txt", "w")
for element in def_lis:
    textfile.write(element + "\n")
textfile.close()
```
